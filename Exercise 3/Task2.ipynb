```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp  # numerical integration solver
import matplotlib.gridspec as gridspec
import cmath
import math
```

```python
### Task 2 - dynamical system "x_dot = alpha - x^2"
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(1, 1, 1)

# x_0 = 0 --> 0 = alpha - x^2
#             x^2 = alpha
#             x = ± sqrt(alpha)

def sqrt_plus(a):
    return sqrt(a)

def sqrt_minus(a):
    return -sqrt(a)


# Given interval in the exercise sheet
interval = linspace(0, 1)

ax1.plot(domain, sqrt_plus(interval), 'b-', label = 'stable state', linewidth = 3)
ax1.plot(domain, sqrt_minus(interval), 'r--', label = 'unstable state', linewidth = 3)

ax1.plot([0], [0], '')
ax1.axis([-1, 1, -1.5, 1.5])
ax1.set_xlabel('alpha')
ax1.set_ylabel('x_dot')
ax1.set_title('Bifurcation Diagram')
```

```python
### Task 2 - dynamical system "x_dot = alpha - 2*x^2 - 3"
```

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(1, 1, 1)

# for x_0 = 0: --> 0 = alpha - 2*x^2 - 3 
#                  2*x^2 = alpha - 3
#                  x^2 = (alpha - 3) / 2
#                  x = ± sqrt((alpha - 3) / 2)

def sqrt_plus_2(a):
    return sqrt((a - 3) / 2)

def sqrt_minus_2(a):
    return -sqrt((a - 3) / 2)


# Given interval in the exercise sheet
interval = linspace(3, 4)

ax1.plot(domain, sqrt_plus_2(interval), 'b-', label = 'stable state', linewidth = 3)
ax1.plot(domain, sqrt_minus_2(interval), 'r--', label = 'unstable state', linewidth = 3)

ax1.plot([0], [0], '')
ax1.axis([-1, 1, -1.5, 1.5])
ax1.set_xlabel('alpha')
ax1.set_ylabel('x_dot')
ax1.set_title('Bifurcation Diagram')
```
