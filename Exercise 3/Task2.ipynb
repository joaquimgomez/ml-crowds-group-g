## Dynamical system 1 $$\dot{x} = \alpha - x^2$$

We have that

$x_0 = 0 \to 0 = \alpha - x^2$

$\to x^2 = \alpha$

$\to x = \pm \sqrt{\alpha}$

```python
def sqrt_plus(a):
    return np.sqrt(a)

def sqrt_minus(a):
    return -np.sqrt(a)

interval = np.linspace(0, 1)
plot_bifurcation_diagram(interval, 
                         [sqrt_plus(interval), sqrt_minus(interval)], 
                         labels = ["stable", "unstable"],
                         axis = [-1, 1, -2, 2])
```

## Dynamical system 2 $$\dot{x} = \alpha - 2x^2 - 3$$

We have that

$x_0 = 0 \to 0 = \alpha - 2 \cdot x^2 - 3$

$\to 2 \cdot x^2 = \alpha - 3$

$\to x^2 = \frac{(\alpha - 3)}{2}$

$\to x = \pm \sqrt{\frac{(\alpha - 3)}{2}}$

```python
def sqrt_plus_2(a):
    return sqrt((a - 3) / 2)

def sqrt_minus_2(a):
    return -sqrt((a - 3) / 2)

interval = linspace(3, 4)
plot_bifurcation_diagram(interval, 
                         [sqrt_plus_2(interval), sqrt_minus_2(interval)], 
                         labels = ["stable", "unstable"],
                         axis = [2, 4, -1, 1])
```
