```python
from utils import *

import os

import matplotlib.pyplot as plt
from matplotlib import colors
```

# Task 4.1 - Static 1000 Visualization

```python
os.chdir("../output/Task4-1-Static1000/")
file_path = os.path.join(os.getcwd(), "SIRInformation.csv")
print(file_path)
```

```python
data = pd.read_csv(file_path, delimiter=" ")

ID_SUSCEPTIBLE = 1
ID_INFECTED = 0
ID_REMOVED = 2

group_counts = file_df_to_count_df(data, ID_INFECTED=ID_INFECTED, ID_SUSCEPTIBLE=ID_SUSCEPTIBLE, ID_REMOVED=ID_REMOVED)
```

```python
infected = group_counts["group-i"]
susceptible = group_counts["group-s"]
recovered = group_counts["group-r"]
simulation_time = group_counts['simTime']

plt.figure(figsize=(12,6))
plt.plot(simulation_time, recovered, color="green")
plt.plot(simulation_time, infected, color="red")
plt.plot(simulation_time, susceptible, color="blue")

plt.xlabel("Simulation Time")
plt.ylabel("Agents")
plt.legend(["Recovered", "Infected", "Susceptible"], loc="lower right" )

plt.title("Task 4.1")

plt.show()
```
