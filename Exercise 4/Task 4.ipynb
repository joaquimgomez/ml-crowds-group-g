```python
from tensorflow import keras
from VAE import *

import matplotlib.pyplot as plt # for plotting model loss
from tensorflow.keras.utils import plot_model # for plotting model diagram

%load_ext autoreload
%autoreload 2
```

Let's load the data and take a look at it:

```python
train = np.load("data/FireEvac_train_set.npy")
test = np.load("data/FireEvac_test_set.npy")
```

```python
plt.plot(train[:,0], train[:,1], 'o');
plt.show()
```

```python
plt.plot(test[:,0], test[:,1], 'o');
plt.show()
```

Let's normalize [-1, 1]

```python
X_train = 2.*(train - np.min(train))/np.ptp(train)-1
X_test = 2.*(test - np.min(test))/np.ptp(test)-1
```

Let's train the VAE model:

```python
original_dim = 2
latent_dim = 64
```

```python
encoder, visible = create_encoder(original_dim=original_dim, latent_dim=latent_dim)
decoder = create_decoder(latent_dim=latent_dim, original_dim=original_dim)
```

```python
(vae, encoder, decoder) = create_vae(encoder, decoder, visible, original_dim)
```

```python
epochs = 1000
batch_size = 1024
```

```python
(vae, encoder, decoder) = train_and_plot_vae(vae,
                                             encoder,
                                             decoder,
                                             X_train,
                                             X_test,
                                             epochs=epochs,
                                             batch_size=batch_size)
```

```python
x_e = encoder.predict(X_test)
x_d = decoder.predict(x_e[2])
x_reconstruct = x_d

plt.plot(X_test[:,0], X_test[:,1], 'o');
plt.show()
plt.plot(x_reconstruct[:,0], x_reconstruct[:,1], 'o');
plt.show()
```
